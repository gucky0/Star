```python
import pandas as pd
import numpy as np

medals_dict = {}

# range end is exclusive
for i in range(2000,2017,4): 
    medals_dict[f'medal_{i}'] = pd.read_html(f"http://en.wikipedia.org/wiki/{i}_Summer_Olympics_medal_table")[2].head(3)
    

    df = medals_dict[f'medal_{i}']

    #create new List and put in Country names (By Index)
    countries = []
    for i in range(3):    
        countries.append(df.loc[i][1])

    #Remove Nation or NOC column
    df = df.drop(df.columns[1], axis=1)

    #Remove other unimportant columns
    df = df.drop(['Gold', 'Silver', 'Bronze'], axis = 1)

    #put in our new list of country names
    df.insert(1, 'Countries', countries, False)

# print(df)
# df.loc[:-1][:-1]
df.loc[:,["Rank","Countries"]]
```
